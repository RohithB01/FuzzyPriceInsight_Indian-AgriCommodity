# **DATA PREPROCESSING**

# **Load the Excel file**

```python
import pandas as pd

Potato = pd.read_excel("/content/Potato_Raw data.xlsx")
Onion =  pd.read_excel("/content/Onion_Raw data.xlsx")
Tomato =  pd.read_excel("/content/Tomato_Raw data.xlsx")
```

# **Standardize column names**

```python
Potato.columns = Potato.columns.str.strip().str.lower().str.replace(' ', '_')
Onion.columns = Onion.columns.str.strip().str.lower().str.replace(' ', '_')
Tomato.columns = Tomato.columns.str.strip().str.lower().str.replace(' ', '_')
```

# **Handle missing values**

```python
Potato.isnull().sum()
Onion.isnull().sum()
Tomato.isnull().sum()
```

```python
Potato = Potato.dropna(how='all')  # Drop rows where all elements are NaN
Potato = Potato.fillna(method='ffill')  # Forward fill

Onion = Onion.dropna(how='all')  # Drop rows where all elements are NaN
Onion = Onion.fillna(method='ffill')  # Forward fill

Tomato = Tomato.dropna(how='all')  # Drop rows where all elements are NaN
Tomato = Tomato.fillna(method='ffill')  # Forward fill
```

# **Convert date columns**

```python
if 'date' in Potato.columns:
    Potato['date'] = pd.to_datetime(Potato['date'], errors='coerce')

if 'date' in Onion.columns:
    Onion['date'] = pd.to_datetime(Onion['date'], errors='coerce')

if 'date' in Tomato.columns:
    Tomato['date'] = pd.to_datetime(Tomato['date'], errors='coerce')
```

# **Drop duplicate rows**

```python
Potato = Potato.drop_duplicates()
Onion = Onion.drop_duplicates()
Tomato = Tomato.drop_duplicates()
```

# **Final preview**

```python
print(Potato.describe())
print(Potato.head())

print(Onion.describe())
print(Onion.head())

print(Tomato.describe())
print(Tomato.head())
```

```python
Potato.to_excel('Potato_Cleaned.xlsx', index = False)
Onion.to_excel('Onion_cleaned.xlsx', index = False)
Tomato.to_excel('Tomato_Cleaned.xlsx', index = False)
```
